```python
def dem(x):
    y = x
    temp_dict = dict()
    temp_dict['line'] = y['line'].tolist()[0]
    temp_dict['construction'] = y['construction'].tolist()[0]
    plat = y['point_latitude'].tolist()
    plat_str = ", ".join(map(str, plat))
    plon = y['point_longitude'].tolist()
    plon_str = ", ".join(map(str, plon))
    temp_dict['point_latitudes'] = plat_str
    temp_dict['point_longitudes'] = plon_str
    df_t = pd.DataFrame.from_records([temp_dict])
    return df_t
    
    
gbobjs = df[:5].groupby(['line', 'construction'])['point_latitude', 'point_longitude', 'line', 'construction']
dff = gbobjs.apply(lambda x : dem(x))
dff = dff.reset_index(drop=True)
dff
```
